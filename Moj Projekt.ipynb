```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('covid.csv')
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
pol = (df['countriesAndTerritories'] == 'Poland')
dfpol = df.loc[pol]

dates = dfpol['dateRep'][::-1]
cases = dfpol['cases'][::-1]
deaths = dfpol['deaths'][::-1]
```

```python
dfpol
```

```python
plt.plot(dates, cases)

plt.title("Square Numbers", fontsize=24)
plt.xlabel("Date")
plt.ylabel("Cases")
#plt.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
plt.plot(dates, deaths)

plt.title("Square Numbers", fontsize=24)
plt.xlabel("Date")
plt.ylabel("Deaths")
#plt.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
slices = [cases.sum(), deaths.sum()]
explode = [0, 0.1]

plt.pie(slices, explode = explode, shadow=True, startangle=0, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title('Cases/Deaths ratio in Poland')
plt.tight_layout()
plt.savefig('plot3.png')
plt.show()
```

```python
plt.bar(dates, cases,
         color = 'y', label='Cases')
plt.plot(dates, deaths)

plt.xlabel('Ages')
plt.ylabel('Median Salary (USD')
plt.title('Median Salary (USD) by Age')

plt.legend()
plt.tight_layout()
plt.show()
```
